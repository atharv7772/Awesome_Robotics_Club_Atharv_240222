```python
import pandas as pd
```

```python
# Task 2.1: Create a DataFrame from the data of students above
data = {
    'Roll No.': [2301, 2310, 2302, 2304, 2306, 2305, 2307, 2312, 2315, 2316, 2320, 2308, 2309, 2314, 2319],
    'Name': ['Alice', 'Smith', 'Bob', 'Charlie', 'Eric', 'Diana', 'John', 'Joey', 'Bean', 'Charlie', 'Jack', 'Jill', 'George', 'Tabitha', 'Phoebe'],
    'Age': [24, None, 22, 23, None, 24, None, 30, 26, 23, 22, 21, 24, 31, 24],
    'Marks': [85, None, 78, 92, 95, 88, None, 64, 72, 92, 83, 69, 89, 100, 91]
}
df = pd.DataFrame(data)
```

***Hint:*** **NaN value** is represented by **None**

```python
# Task 2.2: Inspect the DataFrame(use the most basic functions to know the statistical quantities)
print(df.describe())
print(df.info())
```

```python
# Task 2.3: Select the Marks column and Print it
print(df['Marks'])
```

```python
# Task 2.4: Access the second row
print(df.iloc[1])
```

```python
# (BONUS) Task 2.5: Access only the rows where Marks > 80
print(df[df['Marks'] > 80])
```

```python
# (BONUS, but really fun try to find it do it how :) ) Task 2.6: Make a function to assign grade
# (i) A if marks > 85
# (ii) B if marks > 75
# (iii) C if marks > 65
def assign_grade(marks):
    if marks > 85:
        return 'A'
    elif marks > 75:
        return 'B'
    elif marks > 65:
        return 'C'
    else:
        return 'D'

df['Grade'] = df['Marks'].apply(lambda x: assign_grade(x) if pd.notnull(x) else None)
```

```python
# Task 2.7: Rename Marks column to Score
df.rename(columns={'Marks': 'Score'}, inplace=True)
```

```python
# Task 2.8: Replace NaN in Age with mode and NaN in Marks with mean
df['Age'].fillna(df['Age'].mode()[0], inplace=True)
df['Score'].fillna(df['Score'].mean(), inplace=True)
```

```python
# Task 2.9: Remove duplicates
df.drop_duplicates(inplace=True)
```

```python
# Task 2.10: Group by the data by Grade and print the data
print(df.groupby('Grade').size())
print(df.groupby('Grade').mean(numeric_only=True))
```

```python
# Task 2.11: Sort Data in increasing order of Age
df.sort_values(by='Age', inplace=True)
print(df)
```

```python
# Task 2.12: Export the dataframe to CSV
df.to_csv('students_cleaned.csv', index=False)
```

# ***Part 3: Matplotlib***

```python
import matplotlib.pyplot as plt
```

```python
# Task 3.1: Plot a Pie Chart depicting the percentag of grades that the students have got.
grade_counts = df['Grade'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(grade_counts, labels=grade_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Percentage of Grades')
plt.show()
```

```python
# Task 3.2: Plot a Line Graph of Roll No. vs Marks
plt.figure(figsize=(10, 5))
plt.plot(df['Roll No.'], df['Score'], marker='o', linestyle='-')
plt.title('Roll No. vs Score')
plt.xlabel('Roll No.')
plt.ylabel('Score')
plt.grid(True)
plt.show()
```

```python
# Task 3.3: Plot a Histogram of for showing the Age distribution
plt.figure(figsize=(8, 5))
plt.hist(df['Age'], bins=6, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of Students')
plt.show()
```
